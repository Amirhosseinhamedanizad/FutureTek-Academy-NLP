```python
import requests
from bs4 import BeautifulSoup

# Send an HTTP GET request to the web page
url = "https://www.wolfram.com/mathematica/"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find and extract specific data elements using HTML tags or CSS selectors
title = soup.find("h1").text.strip()
description = soup.find("div", id="overview-product-header").text.strip()
features = soup.find_all("div", class_="feature-row")
benefits = soup.find_all("div", class_="benefit")

# Extract additional information from the website
company = soup.find("div", id="_nav-company").text.strip()
pricing = soup.find("li", id="_product-pricing").text.strip()

# Print the extracted data
print("Title:", title)
print("Description:", description)
print("Features:")
for feature in features:
    feature_title = feature.find("h3").text.strip()
    feature_description = feature.find("p").text.strip()
    print("- Title:", feature_title)
    print("  Description:", feature_description)
    print()

print("Benefits:")
for benefit in benefits:
    benefit_title = benefit.find("h4").text.strip()
    benefit_description = benefit.find("p").text.strip()
    print("- Title:", benefit_title)
    print("  Description:", benefit_description)
    print()

print("Company:", company)
print("Pricing:", pricing)

```
